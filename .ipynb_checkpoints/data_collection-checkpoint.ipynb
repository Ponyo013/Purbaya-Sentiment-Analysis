```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time
import random
import csv

# --- KONFIGURASI ---
user_ig = "masteedarkflame"
pass_ig = "Aviator99"
hashtag = "menterikeuangan"
TARGET_POSTS = 70
COMMENTS_PER_POST = 50 

chrome_options = Options()
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 20)

def smart_click_not_now():
    """Menangani pop-up yang sering muncul setelah login"""
    buttons = [
        "//div[@role='button' and contains(text(), 'Not now')]",
        "//button[contains(text(), 'Not Now')]",
    ]
    for xpath in buttons:
        try:
            element = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, xpath)))
            element.click()
            time.sleep(2)
        except:
            continue
            
def refresh_coordinates():
    print("Sedang membersihkan koordinat scroll...")
    # 1. Scroll ke paling atas dulu
    driver.execute_script("window.scrollTo(0, 0);")
    time.sleep(2)
    
    # 2. Scroll ke bawah secara bertahap (incremental scroll)
    # Ini lebih efektif daripada langsung ke scrollHeight
    for i in range(1, 5):
        pixels = i * 800
        driver.execute_script(f"window.scrollTo(0, {pixels});")
        time.sleep(random.uniform(1.0, 2.0))
    
    # 3. Kembali ke posisi paling bawah untuk memicu data baru
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

try:
    # 1. LOGIN
    driver.get("https://www.instagram.com")
    wait.until(EC.presence_of_element_located((By.NAME, "username"))).send_keys(user_ig)
    wait.until(EC.presence_of_element_located((By.NAME, "password"))).send_keys(pass_ig)
    wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
    
    print("Login berhasil, melewati pop-up...")
    time.sleep(6)
    smart_click_not_now()
    
    # 2. MENGUMPULKAN 100 LINK
    driver.get(f"https://www.instagram.com/explore/tags/{hashtag}/")
    post_links = set()
    print(f"Mencari {TARGET_POSTS} link postingan...")

    last_count = 0
    stuck_counter = 0
    
    while len(post_links) < TARGET_POSTS:
        elements = driver.find_elements(By.XPATH, "//a[contains(@href, '/p/') or contains(@href, '/reels/')]")
        for el in elements:
            link = el.get_attribute("href")
            post_links.add(link)
            if len(post_links) >= TARGET_POSTS: break
    
        # Logika deteksi stuck
        if len(post_links) == last_count:
            stuck_counter += 1
        else:
            last_count = len(post_links)
            stuck_counter = 0

        # Jika 3 kali scroll jumlah link tidak bertambah, bersihkan koordinat
        if stuck_counter >= 3:
            refresh_coordinates()
            stuck_counter = 0 # reset counter setelah koordinat dibersihkan
        else:
            # Scroll normal ke bawah
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(random.uniform(3, 5))
    
        print(f"Total link didapat: {len(post_links)}")

    # 3. SCRAPING KOMENTAR
    # Menyiapkan file CSV
    with open('hasil_scraping_ig.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Link Post', 'Komentar'])
    
        for i, link in enumerate(list(post_links)):
            print(f"\n[{i+1}/{TARGET_POSTS}] Membuka: {link}")
            driver.get(link)
            time.sleep(random.uniform(5, 7))
        
            comments_found = set()
            last_height = 0
            
            # Menggunakan XPath spesifik yang Anda berikan untuk area scroll internal
            xpath_internal_scroll = "/html/body/div[1]/div/div/div[2]/div/div/div[1]/div[1]/div[1]/section/main/div/div[1]/div/div[2]/div/div[2]"
            
            try:
                # Pastikan area komentar muncul
                comment_section = wait.until(EC.presence_of_element_located((By.XPATH, xpath_internal_scroll)))
                print("Area scroll internal ditemukan.")
            except:
                print("Gagal menemukan area scroll internal, mencoba fallback...")
                comment_section = None
    
            # Loop scroll untuk mendapatkan jumlah komentar yang ditargetkan
            for scroll_attempt in range(20): # Maksimal 20 kali scroll per postingan
                try:
                    # 1. Ambil semua elemen komentar (biasanya dalam span dengan class tertentu)
                    # Selektor 'span._ap3a' sering digunakan untuk teks komentar utama
                    spans = driver.find_elements(By.CSS_SELECTOR, "span._ap3a, span.x1lliihq")
                    
                    for s in spans:
                        text = s.text.strip()
                        # Filter sederhana: komentar biasanya punya spasi dan bukan angka saja
                        if text and " " in text and len(text) > 5:
                            comments_found.add(text)
                    
                    print(f"Komentar terkumpul sementara: {len(comments_found)}")
                    
                    if len(comments_found) >= COMMENTS_PER_POST:
                        break
    
                    # 2. Eksekusi Scroll Internal
                    if comment_section:
                        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", comment_section)
                    else:
                        # Fallback jika container tidak ditemukan
                        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                    
                    time.sleep(random.uniform(2, 3))
                    
                    # Cek apakah sudah mentok bawah (infinite scroll check)
                    new_height = driver.execute_script("return arguments[0].scrollHeight", comment_section) if comment_section else 0
                    if new_height == last_height:
                        # Cek tombol "load more" (lingkaran plus)
                        try:
                            load_more_btn = driver.find_element(By.XPATH, "//*[local-name()='svg' and (@aria-label='Load more comments' or @aria-label='Muat komentar lainnya')]/..")
                            driver.execute_script("arguments[0].click();", load_more_btn)
                            time.sleep(2)
                        except:
                            break # Benar-benar habis
                    last_height = new_height
    
                except Exception as e:
                    print(f"Iterasi terhenti: {e}")
                    break
    
            # Simpan hasil dari satu postingan ke CSV
            final_comments = list(comments_found)[:COMMENTS_PER_POST]
            for comment in final_comments:
                writer.writerow([link, comment])
            
            print(f"Berhasil menyimpan {len(final_comments)} komentar dari postingan ini.")

except Exception as e:
    print(f"Error: {e}")

finally:
    print("Proses selesai. Data tersimpan di hasil_scraping_ig.csv")
    driver.quit()
```
